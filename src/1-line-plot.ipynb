### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt  
import numpy as np  

# Generate x values from 0 to 10  
x = np.arange(0, 11, 1)  # generates numbers from 0 to 10  
y = x**2  # calculate the square of x  

# Create the plot  
plt.plot(x, y, marker='o')  

# Customize the plot  
plt.title('Line Plot of y = x^2')  
plt.xlabel('X values')  
plt.ylabel('Y values (y = x^2)')  
plt.grid(True)  

# Show the plot  
plt.show()
```
